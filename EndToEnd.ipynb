### Chia đặc điểm thành tính năng độc lập và phụ thuộc
Independent and dependent features


```python
x = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state = 42  )
```

```python
x_train
```

```python
x_test
```

```python
### Chuẩn hóa dữ liệu
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train) # scaling và học Học cách biến đổi của x_train ( Lấy mean, std )
```

```python
x_test = scaler.transform(x_test) #dùng cách của x_train cho x_test. Không lấy mean,std
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(x_train, y_train)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
# tham so da duoc huan luyen
```

```python
regression.get_params()
```

```python
reg_pred = regression.predict(x_test)
```

```python
reg_pred
```

```python
plt.scatter(y_test,reg_pred)
```

```python
# ve Bieu do phan du
residuals = y_test - reg_pred
residuals
```

```python
sns.displot( residuals , kind = "kde")
```

```python
#  Scatter plot with respect to prediction and residuals
```

```python
plt.scatter(reg_pred, residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
print(mean_absolute_error(y_test, reg_pred))
print(mean_squared_error(y_test, reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

```python
## R bình phương hiệu chỉnh 
## R square and adjusted R square
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,reg_pred)
print(score)
```

```python
1 - (1 - score)* (len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
```
